```python
def draw(arr, x_label, y_label, title, legend, color, file_name=None):
    result, = plt.plot(range(len(arr)), arr, '-o', label=legend, color=color)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    return result
```

```python
def write_file(content, file_name="time.txt"):
    with open(file_name, "w") as f:
        f.write(content+"\n")
```

```python
def dispersion(arr, exp_value):
    return sum([(exp_value - x)**2 for x in arr])/(len(arr)-1)
```

```python
def upload_to_drive(file_name):
    # If modifying these scopes, delete the file token.pickle.
    SCOPES = ['https://www.googleapis.com/auth/drive']

    creds = None
    if os.path.exists('token.pickle'):
        with open('token.pickle', 'rb') as token:
            creds = pickle.load(token)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open('token.pickle', 'wb') as token:
            pickle.dump(creds, token)

    service = build('drive', 'v3', credentials=creds)

    #results = service.files().list(
     #   pageSize=10, fields="nextPageToken, files(id, name)").execute()
    #items = results.get('files', [])
    file_metadata = {'name': file_name}
    media = MediaFileUpload(file_name)
    file = service.files().create(body=file_metadata,
                                    media_body=media,
                                    fields='id').execute()
```

###Lab 1.2 and task.

```python
def correlation_tau(arr, exp_val):
    arr = np.array(arr)
    Rxx = np.zeros(len(arr))
    for tau in range(len(arr)):
        Rxx[tau] = np.sum((arr[:len(arr) - tau] - exp_val) * (arr[tau: len(arr)] - exp_val)) / (len(arr) - 1)
    return Rxx
```

```python
def correlation_tau_xy(arr1, arr2, exp_val1, exp_val2):
    arr1 = np.array(arr1)
    arr2 = np.array(arr2)
    Rxy = np.zeros(len(arr1))
    for tau in range(len(arr1)):
        Rxy[tau] = np.sum((arr1[:len(arr1) - tau] - exp_val1) * (arr2[tau: len(arr2)] - exp_val2)) / (len(arr1) - 1)
    return Rxy
```

```python
def task_lab1_2(signal):
    """
    Method for creating signals in task for 1.2 lab
    """
    N = len(signal)
    signal_1 = np.copy(signal)
    signal_1[N//3:2*N//3] = 0
    signal_2 = np.copy(signal)
    signal_2[0:N//3] = 0
    signal_2[2*N//3:] = 0
```

```python
def lab1_2(signal_1, signal_2):
    """
    Draw corralation for task for 1.2 lab
    """
    x_t = draw(signal_1, "t", "x(t)", "Signal", "X(t)", "blue")
    y_t = draw(signal_2, "t", "x(t)", "Signal", "Y(t)", "orange")
    plt.legend(handles=[x_t, y_t], loc='upper right')
    plt.grid()
    plt.show()
    correlation_1 = correlation_tau(signal_1, np.mean(signal_1))
    Rxx = draw(correlation_1, "tau", "R(tau)", "Correlation", "Rxx(tau)", "red")
    plt.legend(handles=[Rxx])
    plt.grid()
    plt.show()

    correlation_xy = correlation_tau_xy(signal_1, signal_2, np.mean(signal_1), np.mean(signal_2))
    Rxy = draw(correlation_xy, "tau", "Rxy(tau)", "Correlation", "Rxy(tau)", "blue")
    plt.legend(handles=[Rxy])
    plt.grid()
    plt.show()
    correlation_yx = correlation_tau_xy(signal_2, signal_1, np.mean(signal_2), np.mean(signal_1))
    Ryx = draw(correlation_yx, "tau", "Ryx(tau)", "Correlation", "Ryx(tau)", "orange")
    plt.legend(handles=[Ryx])
    plt.grid()
    plt.show()
    print(f"Correlation (Rxy(tau = 0)) (1): {correlation_xy[0]}")
    print(f"Correlation (Ryx(tau = 0)) (2): {correlation_yx[0]}")
```

###Furie tranformations and task for 2.1 lab.

```python
def dft(signal):
    N = len(signal)
    spectr = []
    for p in range(N):
        R, I = 0, 0
        for k in range(N):
            R += signal[k]*cos(2*pi*p*k/N)
            I += signal[k]*sin(2*pi*p*k/N)
        spectr.append(complex(R, -I))
    return spectr
```

```python
def table_dft(signal, half=True):
    """
    Table method task for 2.1 lab and print table shape
    """
    N = len(signal)
    coef_matrix = np.zeros((N, N), dtype=complex)
    #Creation N*N/2 coef table if half == True else - full table. Lower triangular matrix
    for p in range(N):
        for k in range(N):
            if half and k>p:
                break
            coef_matrix[p][k] = complex(cos(2*pi*p*k/N), -sin(2*pi*p*k/N))
    # print table shape
    print("Shape of coef table: ", coef_matrix.shape)
    print("Size (not zero value): ", len(coef_matrix[coef_matrix != 0]))
    spectr = []
    for p in range(N):
        C = complex(0,0)
        for k in range(N):
            if half and k>p:
                C += signal[k]*coef_matrix[k][p]
            else: 
                C += signal[k]*coef_matrix[p][k]
        spectr.append(C)
    return spectr
```

```python
def fft(signal):
    """
    Lab 2.2. FFT
    """
    N = len(signal)
    f1 = []
    f2 = []
    for p in range(N//2):
        c1, c2 = complex(0, 0), complex(0, 0)
        for k in range(N//2):
            c1 += signal[2*k]*complex(cos(4*pi*p*k/N), -sin(4*pi*p*k/N))
            c2 += signal[2*k+1]*complex(cos(4*pi*p*k/N), -sin(4*pi*p*k/N))
        f1.append(c1)
        f2.append(c2)
        
    spectr = []
    for p in range(N):
        Wp = complex(cos(2 * pi * p / N), -sin(2 * pi * p / N))
        index = p if p<N//2 else p-N//2
        spectr.append(f1[index]+f2[index]*Wp)
        
    return spectr
```

```python
def lab2(signal, method):
    """
    Method for draw polar spectr for task for 2.1 lab 
    """
    plot = draw(signal, "t", "x(t)", "Signal", "X(t)", "blue")
    plt.grid()
    plt.show()
    if method == 'dft':
        spectr = dft(signal)
    elif method == 'fft':
        spectr = fft(signal)
    else:
        spectr = table_dft(signal)
    polar_spectr = np.array(list(map(lambda x: cmath.polar(x), spectr)))
    ampl = draw(polar_spectr[:, 0], "t", "A(t)", "Polar Spectr, "+method, "Amplitude", "blue")
    plt.legend(handles=[ampl], loc='upper right')
    plt.grid()
    plt.show()
    phase = draw(polar_spectr[:, 1], "t", "Phi(t)", "Polar Spectr, "+method, "Phase", "red")
    plt.legend(handles=[phase], loc='upper right')
    plt.grid()
    plt.show()
```
